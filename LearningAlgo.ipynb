```python

# set dataset name and run, the output results file and plot are exported to the folder LearningAlgoComparison/  
dataset = "adult"

import csv
import matplotlib.pyplot as plt
import numpy as np


variant = "NoLeafEdgesWithSplitValues"
scoring_function = 'accuracy'
pattern_max_size = 6
filesPath = "forests/rootedFrequentTrees"
filesPath_RF = "arch-forest/data"
resultsPath = "LearningAlgoComparison"




accuracy_list_naive_bayes = []
accuracy_list_log_reg = []
accuracy_list_svm = []
accuracy_list_knn = []
rf_list = []


for rf_depth in (5,10,15,20):
    
    with open(filesPath+'/'+dataset+'/Results_'+variant+'/leq'+str(pattern_max_size)+'/'+'RF_'+str(rf_depth)+'_'+scoring_function+'.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='\n')
        line_count = 0
        for row in csv_reader:
            rowStr = str(row).split(',')
            if line_count < 120:
                accuracy_list_naive_bayes.append(rowStr[1])
                rf_list.append('RF_'+str(rf_depth)+'_'+rowStr[0][2:])
                
                line_count+=1
    csv_file.close()
 

for rf_depth in (5,10,15,20):
    
    with open(filesPath+'/'+dataset+'/Results_'+variant+'/leq'+str(pattern_max_size)+'/'+'RF_'+str(rf_depth)+'_'+scoring_function+'_LogReg.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='\n')
        line_count = 0
        for row in csv_reader:
            rowStr = str(row).split(',')
            if line_count < 120:
                accuracy_list_log_reg.append(rowStr[1])
                #rf_list.append('RF_'+str(rf_depth)+'_t'+str(line_count))
                
                line_count+=1
    csv_file.close()
    
for rf_depth in (5,10,15,20):
    
    with open(filesPath+'/'+dataset+'/Results_'+variant+'/leq'+str(pattern_max_size)+'/'+'RF_'+str(rf_depth)+'_'+scoring_function+'_LinearSVM.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='\n')
        line_count = 0
        for row in csv_reader:
            rowStr = str(row).split(',')
            if line_count < 120:
                accuracy_list_svm.append(rowStr[1])
                #rf_list.append('RF_'+str(rf_depth)+'_t'+str(line_count))
                
                line_count+=1
    csv_file.close()
    
for rf_depth in (5,10,15,20):
    
    with open(filesPath+'/'+dataset+'/Results_'+variant+'/leq'+str(pattern_max_size)+'/'+'RF_'+str(rf_depth)+'_'+scoring_function+'_KNN.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='\n')
        line_count = 0
        for row in csv_reader:
            rowStr = str(row).split(',')
            if line_count < 120:
                accuracy_list_knn.append(rowStr[1])
                #rf_list.append('RF_'+str(rf_depth)+'_t'+str(line_count))
                
                line_count+=1
    csv_file.close()
    
    
    
accuracy_list_dsf = []



accuracy_naive_bayes = np.array(accuracy_list_naive_bayes, dtype=np.float32)
accuracy_log_reg = np.array(accuracy_list_log_reg, dtype=np.float32)
accuracy_svm= np.array(accuracy_list_svm, dtype=np.float32)
accuracy_knn = np.array(accuracy_list_knn, dtype=np.float32)

best_dsf_rf5_nb = 0
best_dsf_rf10_nb = 0
best_dsf_rf15_nb = 0
best_dsf_rf20_nb = 0
best_dsf_rf5_index_nb = 0
best_dsf_rf10_index_nb = 0
best_dsf_rf15_index_nb = 0
best_dsf_rf20_index_nb = 0

best_dsf_rf5_logreg= 0
best_dsf_rf10_logreg = 0
best_dsf_rf15_logreg = 0
best_dsf_rf20_logreg = 0
best_dsf_rf5_index_logreg = 0
best_dsf_rf10_index_logreg = 0
best_dsf_rf15_index_logreg = 0
best_dsf_rf20_index_logreg = 0

best_dsf_rf5_svm = 0
best_dsf_rf10_svm = 0
best_dsf_rf15_svm = 0
best_dsf_rf20_svm = 0
best_dsf_rf5_index_svm = 0
best_dsf_rf10_index_svm = 0
best_dsf_rf15_index_svm = 0
best_dsf_rf20_index_svm = 0

best_dsf_rf5_knn = 0
best_dsf_rf10_knn = 0
best_dsf_rf15_knn = 0
best_dsf_rf20_knn = 0
best_dsf_rf5_index_knn = 0
best_dsf_rf10_index_knn = 0
best_dsf_rf15_index_knn = 0
best_dsf_rf20_index_knn = 0

for i in range(0,len(accuracy_list_naive_bayes)):
    if (i < 120):
        if (accuracy_naive_bayes[i] > best_dsf_rf5_nb):
            best_dsf_rf5_nb = accuracy_naive_bayes[i]
            best_dsf_rf5_index_nb = i
        if (accuracy_log_reg[i] > best_dsf_rf5_logreg):
            best_dsf_rf5_logreg = accuracy_log_reg[i]
            best_dsf_rf5_index_logreg = i 
        if (accuracy_svm[i] > best_dsf_rf5_svm):
            best_dsf_rf5_svm = accuracy_svm[i]
            best_dsf_rf5_index_svm = i 
        if (accuracy_knn[i] > best_dsf_rf5_knn):
            best_dsf_rf5_knn = accuracy_knn[i]
            best_dsf_rf5_index_knn = i     
            
    if (i >= 120 and i < 240):
        if (accuracy_naive_bayes[i] > best_dsf_rf10_nb):
            best_dsf_rf10_nb = accuracy_naive_bayes[i]
            best_dsf_rf10_index_nb = i
        if (accuracy_log_reg[i] > best_dsf_rf10_logreg):
            best_dsf_rf10_logreg = accuracy_log_reg[i]
            best_dsf_rf10_index_logreg = i
        if (accuracy_svm[i] > best_dsf_rf10_svm):
            best_dsf_rf10_svm = accuracy_svm[i]
            best_dsf_rf10_index_svm = i 
        if (accuracy_knn[i] > best_dsf_rf10_knn):
            best_dsf_rf10_knn = accuracy_knn[i]
            best_dsf_rf10_index_knn = i    
            
    if (i >= 240 and i < 360):
        if (accuracy_naive_bayes[i] > best_dsf_rf15_nb):
            best_dsf_rf15_nb = accuracy_naive_bayes[i]
            best_dsf_rf15_index_nb = i
        if (accuracy_log_reg[i] > best_dsf_rf15_logreg):
            best_dsf_rf15_logreg = accuracy_log_reg[i]
            best_dsf_rf15_index_logreg = i
        if (accuracy_svm[i] > best_dsf_rf15_svm):
            best_dsf_rf15_svm = accuracy_svm[i]
            best_dsf_rf15_index_svm = i 
        if (accuracy_knn[i] > best_dsf_rf15_knn):
            best_dsf_rf15_knn = accuracy_knn[i]
            best_dsf_rf15_index_knn = i    
            
            
    if (i >= 360 and i < 480):
        if (accuracy_naive_bayes[i] > best_dsf_rf20_nb):
            best_dsf_rf20_nb = accuracy_naive_bayes[i]
            best_dsf_rf20_index_nb = i 
        if (accuracy_log_reg[i] > best_dsf_rf20_logreg):
            best_dsf_rf20_logreg = accuracy_log_reg[i]
            best_dsf_rf20_index_logreg = i  
        if (accuracy_svm[i] > best_dsf_rf20_svm):
            best_dsf_rf20_svm = accuracy_svm[i]
            best_dsf_rf20_index_svm = i 
        if (accuracy_knn[i] > best_dsf_rf20_knn):
            best_dsf_rf20_knn = accuracy_knn[i]
            best_dsf_rf20_index_knn = i      
    

accuracy_dsf_best_nb = []
accuracy_dsf_best_nb.append(best_dsf_rf5_nb)
accuracy_dsf_best_nb.append(best_dsf_rf10_nb)
accuracy_dsf_best_nb.append(best_dsf_rf15_nb)
accuracy_dsf_best_nb.append(best_dsf_rf20_nb)

accuracy_dsf_best_logreg = []
accuracy_dsf_best_logreg.append(best_dsf_rf5_logreg)
accuracy_dsf_best_logreg.append(best_dsf_rf10_logreg)
accuracy_dsf_best_logreg.append(best_dsf_rf15_logreg)
accuracy_dsf_best_logreg.append(best_dsf_rf20_logreg)

accuracy_dsf_best_svm = []
accuracy_dsf_best_svm.append(best_dsf_rf5_svm)
accuracy_dsf_best_svm.append(best_dsf_rf10_svm)
accuracy_dsf_best_svm.append(best_dsf_rf15_svm)
accuracy_dsf_best_svm.append(best_dsf_rf20_svm)

accuracy_dsf_best_knn = []
accuracy_dsf_best_knn.append(best_dsf_rf5_knn)
accuracy_dsf_best_knn.append(best_dsf_rf10_knn)
accuracy_dsf_best_knn.append(best_dsf_rf15_knn)
accuracy_dsf_best_knn.append(best_dsf_rf20_knn)

best_dsf_index_nb = np.argmax(accuracy_dsf_best_nb)
best_dsf_nb = np.max(accuracy_dsf_best_nb)

best_dsf_index_logreg = np.argmax(accuracy_dsf_best_logreg)
best_dsf_logreg = np.max(accuracy_dsf_best_logreg)

best_dsf_index_svm = np.argmax(accuracy_dsf_best_svm)
best_dsf_svm = np.max(accuracy_dsf_best_svm)

best_dsf_index_knn = np.argmax(accuracy_dsf_best_knn)
best_dsf_knn = np.max(accuracy_dsf_best_knn)


def convert_index(index):
    new_index = ((index % 120) // 5) + 2
    index_string = str(new_index)
    
    pruning = index % 5
    if pruning == 0:
        output = 't' + index_string 
    elif pruning == 1:
        output = 'pruned_with_sigma_0_0_t' + index_string  
    elif pruning == 2:
        output = 'pruned_with_sigma_0_1_t' + index_string  
    elif pruning == 3:
        output = 'pruned_with_sigma_0_2_t' + index_string  
    elif pruning == 4:
        output = 'pruned_with_sigma_0_3_t' + index_string 
    
    return output


f= open(resultsPath+'/'+dataset+'/'+'best_accuracy_'+dataset+'_comparison'+'.csv',"w")
f.write('Naive Bayes,\n')
f.write('RF,accuracy,\n')
f.write('RF_5_'+convert_index(best_dsf_rf5_index_nb)+','+str(accuracy_dsf_best_nb[0])+',\n')
f.write('RF_10_'+convert_index(best_dsf_rf10_index_nb)+','+str(accuracy_dsf_best_nb[1])+',\n')
f.write('RF_15_'+convert_index(best_dsf_rf15_index_nb)+','+str(accuracy_dsf_best_nb[2])+',\n')
f.write('RF_20_'+convert_index(best_dsf_rf20_index_nb)+','+str(accuracy_dsf_best_nb[3])+',\n')
f.write('\n')
f.write('Best DSF'+','+str(best_dsf_nb)+',\n')
f.write('\n')
f.write('\n')
f.write('Log Reg,\n')
f.write('RF,accuracy,\n')
f.write('RF_5_'+convert_index(best_dsf_rf5_index_logreg)+','+str(accuracy_dsf_best_logreg[0])+',\n')
f.write('RF_10_'+convert_index(best_dsf_rf10_index_logreg)+','+str(accuracy_dsf_best_logreg[1])+',\n')
f.write('RF_15_'+convert_index(best_dsf_rf15_index_logreg)+','+str(accuracy_dsf_best_logreg[2])+',\n')
f.write('RF_20_'+convert_index(best_dsf_rf20_index_logreg)+','+str(accuracy_dsf_best_logreg[3])+',\n')
f.write('\n')
f.write('Best DSF'+','+str(best_dsf_logreg)+',\n')
f.write('\n')
f.write('\n')
f.write('Linear SVM,\n')
f.write('RF,accuracy,\n')
f.write('RF_5_'+convert_index(best_dsf_rf5_index_svm)+','+str(accuracy_dsf_best_svm[0])+',\n')
f.write('RF_10_'+convert_index(best_dsf_rf10_index_svm)+','+str(accuracy_dsf_best_svm[1])+',\n')
f.write('RF_15_'+convert_index(best_dsf_rf15_index_svm)+','+str(accuracy_dsf_best_svm[2])+',\n')
f.write('RF_20_'+convert_index(best_dsf_rf20_index_svm)+','+str(accuracy_dsf_best_svm[3])+',\n')
f.write('\n')
f.write('Best DSF'+','+str(best_dsf_svm)+',\n')
f.write('\n')
f.write('\n')
f.write('KNN,\n')
f.write('RF,accuracy,\n')
f.write('RF_5_'+convert_index(best_dsf_rf5_index_knn)+','+str(accuracy_dsf_best_knn[0])+',\n')
f.write('RF_10_'+convert_index(best_dsf_rf10_index_knn)+','+str(accuracy_dsf_best_knn[1])+',\n')
f.write('RF_15_'+convert_index(best_dsf_rf15_index_knn)+','+str(accuracy_dsf_best_knn[2])+',\n')
f.write('RF_20_'+convert_index(best_dsf_rf20_index_knn)+','+str(accuracy_dsf_best_knn[3])+',\n')
f.write('\n')
f.write('Best DSF'+','+str(best_dsf_knn)+',\n')
f.write('\n')
f.write('\n')


f.close()

```
