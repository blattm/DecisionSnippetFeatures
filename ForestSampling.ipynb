```python
import json
import random

variant = "NoLeafEdges"
pattern_max_size = 6
jsonPath = "../frequentTreesInRandomForests/forests/rootedFrequentTrees/"
patternsPath = "../frequentTreesInRandomForests/InferenceComparison"
patternList=[]


def sampleRandomForest(dataset, samples_count, patternsCount, max_depth):    

    with open(jsonPath+dataset+'/NoLeafEdges/leq6/'+'RF_'+str(max_depth)+'_t1.json') as json_file:
         data = json.load(json_file)
         for c in range(0,samples_count):
                
                patternList=[]
                for i in range(0,patternsCount-1):
                   print(str(data[random.randint(0,len(data)-1)]).replace("'", '"'))
                   patternList.append(str(data[random.randint(0,len(data)-1)]).replace("'", '"')+',')
                patternList.append(str(data[random.randint(0,len(data)-1)]).replace("'", '"')+']')
                f= open(jsonPath+dataset+'/NoLeafEdges/leq6/Samples/RF_'+str(max_depth)+'/'+'RF_'+str(max_depth)+'_t1_'+str(c+1)+'.json',"w")
                f.write('[')
                for pattern in patternList:
                    
                    f.write(str(pattern)+'\n')
                   
                f.close()            
        
            
    json_file.close()

    
    
dataset = "adult"
samples_count = 10
patternsCount = 134
max_depth = 5   
sampleRandomForest(dataset,samples_count, patternsCount, max_depth)

    
    
```
