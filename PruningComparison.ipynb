```python
import numpy as np


frequentTreesPath = "forests/rootedFrequentTrees/"
resultsPath = "PruningComparison/"

dataset='adult'
variant = 'NoLeafEdgesWithSplitValues'

#complete path to files (with example file):
#forests/rootedFrequentTrees/adult/NoLeafEdgesWithSplitValues/RF_5_pruned_with_sigma_0_1_t2.patterns

for rf_depth in (5,10,15,20):
    
    files_path = frequentTreesPath + dataset + '/' + variant + '/'
    file1 = files_path + 'RF_' + str(rf_depth) + '_t2.patterns'
    file2 = files_path + 'RF_' + str(rf_depth) + '_pruned_with_sigma_0_0_t2.patterns'
    file3 = files_path + 'RF_' + str(rf_depth) + '_pruned_with_sigma_0_1_t2.patterns'
    file4 = files_path + 'RF_' + str(rf_depth) + '_pruned_with_sigma_0_2_t2.patterns'
    file5 = files_path + 'RF_' + str(rf_depth) + '_pruned_with_sigma_0_3_t2.patterns'

    with open(file1, "r") as f:
        RF = f.readlines()

    with open(file2, "r") as f:
        RF_pruned_0_0 = f.readlines()
    
    with open(file3, "r") as f:
        RF_pruned_0_1 = f.readlines()
            
    with open(file4, "r") as f:
        RF_pruned_0_2 = f.readlines()
    
    with open(file5, "r") as f:
        RF_pruned_0_3 = f.readlines()
            

    patterns_RF = []
    patterns_pruned_0_0 = []
    patterns_pruned_0_1 = []
    patterns_pruned_0_2 = []
    patterns_pruned_0_3 = []
    frequencies_RF = []
    frequencies_pruned_0_0 = []
    frequencies_pruned_0_1 = []
    frequencies_pruned_0_2 = []
    frequencies_pruned_0_3 = []

    for line in RF:
        parts = line.split()
        patterns_RF.append( ' '.join(parts[2:]) )
        frequencies_RF.append( int(parts[0]) )
    
    for line in RF_pruned_0_0:
        parts = line.split()
        patterns_pruned_0_0.append( ' '.join(parts[2:]) )
        frequencies_pruned_0_0.append( int(parts[0]) )
    
    for line in RF_pruned_0_1:
        parts = line.split()
        patterns_pruned_0_1.append( ' '.join(parts[2:]) )
        frequencies_pruned_0_1.append( int(parts[0]) )
    
    for line in RF_pruned_0_2:
        parts = line.split()
        patterns_pruned_0_2.append( ' '.join(parts[2:]) )
        frequencies_pruned_0_2.append( int(parts[0]) )
    
    for line in RF_pruned_0_3:
        parts = line.split()
        patterns_pruned_0_3.append( ' '.join(parts[2:]) )
        frequencies_pruned_0_3.append( int(parts[0]) )
    
    RF_depth = 'RF_'+str(rf_depth)
    output_list = []
    output_list.append(' , ' + RF_depth + ', ' + RF_depth + '_pruned_with_sigma_0_0, ' + RF_depth + '_pruned_with_sigma_0_1, ' \
                      + RF_depth + '_pruned_with_sigma_0_2, ' + RF_depth + '_pruned_with_sigma_0_3')
    
    #number of frequent subtrees
    n_RF = len(patterns_RF)
    n_RF_0 = len(patterns_pruned_0_0)
    n_RF_1 = len(patterns_pruned_0_1)
    n_RF_2 = len(patterns_pruned_0_2)
    n_RF_3 = len(patterns_pruned_0_3)
    
    output_list.append('number of frequent subtrees (FS), ' + str(n_RF) + ', ' + str(n_RF_0) + ', ' \
                      + str(n_RF_1) + ', ' + str(n_RF_2) + ', ' + str(n_RF_3) )
    output_list.append('ratio of FS compared to ' + RF_depth + ', 1, ' + str(round(n_RF_0/n_RF,2)) + ', ' \
                      + str(round(n_RF_1/n_RF,2)) + ', ' + str(round(n_RF_2/n_RF,2)) + ', ' + str(round(n_RF_3/n_RF,2)) )
    output_list.append('average frequency of all FS, ' + str(round(np.mean(frequencies_RF),2)) + ', ' + str(round(np.mean(frequencies_pruned_0_0),2)) + ', ' \
                      + str(round(np.mean(frequencies_pruned_0_1),2)) + ', ' + str(round(np.mean(frequencies_pruned_0_2),2)) + ', ' + str(round(np.mean(frequencies_pruned_0_3),2)) )
    
    #look at frequent subtrees that appear in the original RF and also in one of the pruned versions.
    #how many such FS are there and are they "more frequent" in the pruned versions?
    
    equal_list_0 = []
    equal_list_1 = []
    equal_list_2 = []
    equal_list_3 = []
    
    for i, fs in enumerate(patterns_RF):
        try:
            equal_list_0.append([i, patterns_pruned_0_0.index(fs)])
        except ValueError:
            pass
        
        try:
            equal_list_1.append([i, patterns_pruned_0_1.index(fs)])
        except ValueError:
            pass
        
        try:
            equal_list_2.append([i, patterns_pruned_0_2.index(fs)])
        except ValueError:
            pass
        
        try:
            equal_list_3.append([i, patterns_pruned_0_3.index(fs)])
        except ValueError:
            pass
        
    output_list.append('number of FS that also appear in '+ RF_depth + ', ' + str(n_RF) + ', ' + str(len(equal_list_0)) + ', ' \
                      + str(len(equal_list_1)) + ', ' + str(len(equal_list_2)) + ', ' + str(len(equal_list_3)) )
    
    #are the found FS "more frequent" in the pruned versions?
    # -> calculate the average of the frequency-ratios (= frequency in the pruned version divided by the frequency in the original RF)
    mean_ratio_0 = 0
    mean_ratio_1 = 0
    mean_ratio_2 = 0
    mean_ratio_3 = 0
    
    if len(equal_list_0) > 0:
        ratios = [ int(frequencies_pruned_0_0[l[1]]) / int(frequencies_RF[l[0]]) for l in equal_list_0 ]
        mean_ratio_0 = round(np.mean(ratios),2)
    
    if len(equal_list_1) > 0:
        ratios = [ int(frequencies_pruned_0_1[l[1]]) / int(frequencies_RF[l[0]]) for l in equal_list_1 ]
        mean_ratio_1 = round(np.mean(ratios),2)
        
    if len(equal_list_2) > 0:
        ratios = [ int(frequencies_pruned_0_2[l[1]]) / int(frequencies_RF[l[0]]) for l in equal_list_2 ]
        mean_ratio_2 = round(np.mean(ratios),2)
        
    if len(equal_list_3) > 0:
        ratios = [ int(frequencies_pruned_0_3[l[1]]) / int(frequencies_RF[l[0]]) for l in equal_list_3 ]
        mean_ratio_3 = round(np.mean(ratios),2)
        
    output_list.append('average frequency ratio of such FS, 1, ' + str(mean_ratio_0) + ', ' \
                    + str(mean_ratio_1) + ', ' + str(mean_ratio_2) + ', ' + str(mean_ratio_3) )
    
    #save the outputs in a file
    output_file = resultsPath + dataset + '/' + variant + '/' + RF_depth + '_pruning_comparison.csv'
    f = open(output_file, 'a')
    for line in output_list:
        f.write(line)
    f.close()
    
    
    
```
